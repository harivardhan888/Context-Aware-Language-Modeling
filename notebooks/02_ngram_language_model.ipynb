```python
import sys
import os
sys.path.append(os.path.abspath('../src'))

from models import NGramModel
from preprocessing import load_data, clean_text, tokenize

with open('../data/processed/corpus.txt', 'r', encoding='utf-8') as f:
    corpus = f.read().split()

split_idx = int(0.8 * len(corpus))
train_data = corpus[:split_idx]
test_data = corpus[split_idx:]

print(f'Train size: {len(train_data)}')
print(f'Test size: {len(test_data)}')
```

```python
# Train Bigram
bigram_model = NGramModel(n=2)
bigram_model.train(train_data)

# Perplexity
ppl_bi = bigram_model.calculate_perplexity(test_data[:1000]) # Evaluate on subset for speed
print(f'Bigram Perplexity: {ppl_bi}')
```

```python
# Train Trigram
trigram_model = NGramModel(n=3)
trigram_model.train(train_data)

ppl_tri = trigram_model.calculate_perplexity(test_data[:1000])
print(f'Trigram Perplexity: {ppl_tri}')
```
